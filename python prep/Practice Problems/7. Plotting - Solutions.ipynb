2. Write a function that takes in $n$ then plots a random walk of length $n$. You may want to give the user the ability to choose the color and `linewidth` of their walk.

3. Load the `iris.csv` data using `pandas`. Make a 2 x 2 subplot. In the (0,0) subplot plot the `sepal_length` on the $x$ and the `sepal_width` on the $y$, in the (1,0) subplot plot the `petal_length` on the $x$ and the `sepal_width` on the $y$, and finally in the (0,1) subplot plot the `sepal_length` on the $x$ and the `petal_width` on the $y$. Color each point by its `target`. Write the text "Iris Plot" in the (1,1) subplot.

```python
## 1. ANSWER
import numpy as np
```

```python
## 1. ANSWER continued

plt.figure(figsize=(8,6))

plt.hist(np.random.randn(100), color = "red", label="Draw 1")
plt.hist(np.random.randn(100) + 2, color = "blue", label="Draw 2")

plt.legend(fontsize=14)

plt.show()
```

```python
## 2. ANSWER

# from notebook 5 solution
def random_walk(N):
    # get the flips
    flips = np.random.binomial(1,p=.5,size=N)
    flips = 2*flips - 1

    # Get the positions
    pos = np.zeros(N+1)
    pos[1:] = np.cumsum(flips)
    
    # return the positions
    return pos

def plot_walk(N, color="blue", linewidth=1.5):
    # get the walk
    walk = random_walk(N)
    
    # make the figure
    plt.figure(figsize=(8,6))
    
    # need to add 1 to N for range to account for the 
    # 0 position
    plt.plot(range(N+1), walk, linewidth = linewidth, color = color)
    
    plt.xlabel("Step", fontsize=16)
    plt.ylabel("Position", fontsize=16)
    
    plt.title("A Random Walk", fontsize=20)
    
    plt.show()
```

```python
## 2. ANSWER cont.

plot_walk(100)

plot_walk(1000, linewidth=1, color='orange')
```

```python
## 3. ANSWER
iris = pd.read_csv("iris.csv")


iris.head()
```

```python
## 3. ANSWER cont.

fig, ax = plt.subplots(2,2, figsize=(12,12))

ax[0,0].scatter(iris.loc[iris.target==0].sepal_length, 
                iris.loc[iris.target==0].sepal_width,
               c='r', label="setosa")
ax[0,0].scatter(iris.loc[iris.target==1].sepal_length, 
                iris.loc[iris.target==1].sepal_width,
               c='b', label="versicolor")
ax[0,0].scatter(iris.loc[iris.target==2].sepal_length, 
                iris.loc[iris.target==2].sepal_width,
               c='g', label="virginica")
ax[0,0].set_ylabel("Sepal Width", fontsize=16)
ax[0,0].set_xlabel("Sepal Length", fontsize=16)
ax[0,0].legend(fontsize=12)

ax[1,0].scatter(iris.loc[iris.target==0].petal_length, 
                iris.loc[iris.target==0].sepal_width,
               c='r', label="setosa")
ax[1,0].scatter(iris.loc[iris.target==1].petal_length, 
                iris.loc[iris.target==1].sepal_width,
               c='b', label="versicolor")
ax[1,0].scatter(iris.loc[iris.target==2].petal_length, 
                iris.loc[iris.target==2].sepal_width,
               c='g', label="virginica")
ax[1,0].set_ylabel("Sepal Width", fontsize=16)
ax[1,0].set_xlabel("Petal Length", fontsize=16)
ax[1,0].legend(fontsize=12)

ax[0,1].scatter(iris.loc[iris.target==0].sepal_length, 
                iris.loc[iris.target==0].petal_width,
               c='r', label="setosa")
ax[0,1].scatter(iris.loc[iris.target==1].sepal_length, 
                iris.loc[iris.target==1].petal_width,
               c='b', label="versicolor")
ax[0,1].scatter(iris.loc[iris.target==2].sepal_length, 
                iris.loc[iris.target==2].petal_width,
               c='g', label="virginica")
ax[0,1].set_ylabel("Petal width", fontsize=16)
ax[0,1].set_xlabel("Sepal Length", fontsize=16)
ax[0,1].legend(fontsize=12)

ax[1,1].text(0.4,0.45, "Iris", fontsize=20)

plt.show()
```
